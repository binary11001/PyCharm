```python
path = "D:\Capnxt_Energon\energon_month\STOCK CARD EPPL REPORT Mar 2023 (1).xlsx"
# path = "/Users/nithin_sai/Downloads/excels/EPPL REPORT Jan 2022.xlsx"
mar_2023=pd.read_excel(path,header=None)
mar_2023=mar_2023.drop([0,1,2,4,5])
mar_2023 = mar_2023.drop(mar_2023.columns[[0,2,3]],axis = 1)
mar_2023 = mar_2023.dropna(subset=[1])

month=mar_2023.iloc[:,0:32]
month.at[0,1]="Date"

month

month=month.transpose()

new_header = month.iloc[0] #grab the first row for the header
month = month[1:] #take the data less the header row
month.columns = new_header #set the header row as the df header

sqc = month.filter(regex='SQC')
sqc
sqc=sqc.drop(columns=['SQC'])

l=['No. of SQC 14.2 KG',
'SQC in terms of cylinders 14.2 KG',
'Rejected Cyls in SQC 14.2 KG',
 'No. of SQC 19 KG',
 'SQC in terms of cylinders 19 KG',
 'Rejected Cyls in SQC 19 KG',
 'No. of SQC 35 KG',
 'SQC in terms of cylinders 35 KG',
 'Rejected Cyls in SQC 35 KG',
 'No. of SQC 47.5',
 'SQC in terms of cylinders 47.5 KG',
 'Rejected Cyls in SQC 47.5 KG',
 'No. of SQC 425 KG',
 'SQC in terms of cylinders 425 KG',
 'Rejected Cyls in SQC 425 KG',
 'No. of SQC 5 KG (RED)',
 'SQC in terms of cylinders 5 KG (RED)',
 'Rejected Cyls in SQC 5 KG (RED)',
 'No. of SQC 5 KG (BLUE)',
 'SQC in terms of cylinders 5 KG (BLUE)',
 'Rejected Cyls in SQC 5 KG (BLUE)',
 'No. of SQC 5 KG (FTL)',
 'SQC in terms of cylinders 5 KG (FTL)',
 'Rejected Cyls in SQC 5 KG (FTL)',
 'No. of SQC 5 KG (2 KG)',
 'SQC in terms of cylinders 5 KG (2 KG)',
 'Rejected Cyls in SQC 5 KG (2 KG)']

sqc.columns = l
sqc

mar_2023=pd.read_excel(path,header=None)
mar_2023=mar_2023.drop([0,1,2,4,5])
mar_2023 = mar_2023.drop(mar_2023.columns[[0,2,3]],axis = 1)
mar_2023=mar_2023.reset_index()
mar_2023.drop(columns=['index'],inplace=True)
mar_2023 = mar_2023.dropna(subset=[1])

month=mar_2023.iloc[:,0:32]
month.at[0,1]="Date"

for p in range(6,len(month)):
    if p in month.index:
        res=re.sub(' +'," ",month.at[p,1])
        month.at[p,1]=res


# day's table
day={}
j=0
frames=[]

for k in range(4,35):
    day[month.iloc[0][1]]=month.iloc[0][k]

    for i in range(1,len(month)):
        cat=month.iloc[i][1]
        if "14.2" in cat.lower():
            day[month.iloc[i][1]]=month.iloc[i][k]

    frames.append(pd.DataFrame(day,index=[j]))
    j+=1


df_mar_2023=pd.concat(frames)
df_mar_2023.rename(columns={"Day's Production 5 kg FTL/NDM (BLUE)":"Day's Production 5 kg FTL/FTR (BLUE)"},inplace=True)



df_mar_2023 = df_mar_2023[df_mar_2023.columns.drop(list(df_mar_2023.filter(regex='SQC')))]
df_mar_2023


for i in sqc.columns:
    l=list(sqc[i])
    df_mar_2023[i]=l
df_mar_2023=df_mar_2023[:27]
df_mar_2023
# pd.to_datetime(df_mar_2023.iloc[:,0])
```
